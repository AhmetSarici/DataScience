```python
from sklearn.linear_model import LinearRegression
```

```python
linear_reg=LinearRegression()
```

```python
x=data.deneyim
y=data.maas
```

```python
type(x)
```

Numpy'a dönüştürülmesi

```python
x=x.values
y=y.values
```

```python
type(x)
```

```python
x.shape
```

14,1 boyutunda olması gerekir

```python
x=x.reshape(-1,1)
y=y.reshape(-1,1)
```

```python
x.shape
```

Mavi noktalar ayarlanmış oldu, fit işlemi uygulanır yani kırmızı doğru mavi noktalara min(MSE) ile ayarlanması işlemi

```python
linear_reg.fit(x,y)
```

```python
linear_reg.predict([[0]])
```

Y eksenini kestiği nokta 

```python
b0_ = linear_reg.intercept_
b0_
```

Eğimi nedir

```python
b1=linear_reg.coef_

b1
```
