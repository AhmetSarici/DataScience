# Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline 
```

```python
yas=[10,20,30,30,30,40,50,60,70,75]
kilo=[20,60,80,85,86,87,70,90,95,90]
```

```python
npyas=np.array(yas)
npkilo=np.array(kilo)
```

```python
plt.plot(npyas,npkilo,"g");
```

```python
plt.plot(npyas,npkilo,"g");
plt.title("Kilonun Yaşa Göre Değişim Grafiği")
plt.xlabel("Yaş")
plt.ylabel("Kilo")
plt.show()
```

Özelleştirme

```python
npdizi=np.linspace(0,10,20)
npdizi 
```

```python
npdizi2=npdizi**3
npdizi2
```

```python
plt.plot(npdizi,npdizi2,"g*-");
```

```python
#iki grafik
plt.subplot(1,2,1)  #sıra,kolon,sıra
plt.plot(npdizi,npdizi2,"r*-");
plt.subplot(1,2,2)  #sıra,kolon,sıra
plt.plot(npdizi2,npdizi,"g--"); 
```

```python
bfigure = plt.figure()
 #figüre eksen ekleme
figureaxes =bfigure.add_axes([0.2,0.2,0.9,0.9])  #boyut ayarları
figureaxes.plot(npdizi,npdizi2,"g")
figureaxes.set_xlabel("X Ekseni")
figureaxes.set_ylabel("Y Ekseni")
figureaxes.set_title("Grafik Başlığı")
```

```python
fig2=plt.figure()
eksen1=fig2.add_axes([0.1,0.1,0.7,0.7])
eksen2=fig2.add_axes([0.2,0.4,0.3,0.3])

eksen1.plot(npdizi,npdizi2,"g")
eksen1.set_label("X Ekseni")
eksen1.set_label("Y Ekseni")
eksen1.set_title("Ana Grafik Başlığı")

eksen2.plot(npdizi2,npdizi,"b")
eksen2.set_label("X Ekseni")
eksen2.set_label("Y Ekseni")
eksen2.set_title("Küçük Başlık")
```

```python
(bnmfigure,bnmeksen)=plt.subplots(nrows=1,ncols=2)


for eks in bnmeksen:
    eks.plot(npdizi,npdizi2,"g")
    eks.set_xlabel("X Ekseni")
plt.tight_layout() #ikisi arası mesafe    
```

```python
ynifigure=plt.figure(figsize=(5,5), dpi=100) #dpi çözünürlük, figsize ölçek
ynieksen=ynifigure.add_axes([0.1,0.1,0.9,0.9])

ynieksen.plot(npdizi,npdizi**2, label="NumpyDizisi**2")
ynieksen.plot(npdizi,npdizi**3, label="NumpyDizisi**3")
ynieksen.legend(loc=6);
```

```python
#Grafiği kaydetme
ynifigure.savefig("bnm.png",dpi=200)
```

```python
npd1=np.linspace(0,10,20)
npd1
```

```python
npd2=npd1**2
```

```python
(bf1,be1)= plt.subplots()
be1.plot(npd1,npd2,"g", alpha=0.3)

be1.plot(npd2,npd1,color="#3A95A8")
```

```python
(bf2,be2)= plt.subplots()
be2.plot(npd1,npd1+2,"b", alpha=0.3, linewidth=5.0, marker="*", markerfacecolor="y", markersize=15)

be2.plot(npd1,npd1+3,color="#3A95A8",linestyle="-.")

be2.plot(npd1,npd1+4, linestyle="--")
```
