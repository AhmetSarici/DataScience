# Scatter Plot Matrisi
Sayısal değişkenler arasındaki ilişkiyi incelenmesinde rol alır

```python
import seaborn as sns;
iris = sns.load_dataset("iris")
df=iris.copy()
df.head()
```

```python
df.dtypes
```

```python
df.info()
```

```python
sns.pairplot(df);
```

Scatter plot saçılım grafik : Grafik toz bulutu şeklinde veya yapısal bir form yoksa iki değişken arasında bir ilişki yoktur.
Gözlem noktaları incelendiğinde  köşe noktalarda kümeleme oluşuyorsa çaprazlamalarda farklı alt grupları yapılanmaları ifade ediyordur.



---



Kategorik değişkenin eklenmesi

```python
sns.pairplot(df, hue="species");
```

```python
#Farklı işaretler kullanımı markers
sns.pairplot(df, hue="species", markers=["o","s","D"]);
```



---
Grafiğe doğru eklenmesi


```python
sns.pairplot(df, kind="reg", hue="species");
```

#  Isı Heritası (Heatmap)
Veri Seti Hikayesi

```python
import seaborn as sns
flights = sns.load_dataset("flights")
df=flights.copy()
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df["passengers"].describe()
```

Pivot işlemi birinci girdi x ekseni, ikinci y ekseni

```python
#X=Month, y=year, sürekli değişken
df=df.pivot("month","year","passengers");
df
```

```python
sns.heatmap(df);
```

```python
sns.heatmap(df, annot=True, fmt="d");
```

```python
sns.heatmap(df, annot=True, fmt="d", linewidths=.5, linecolor="black");
```

# Çizgi Grafik
* Zamana bağlı değişkenlerin olduğu veri setlerinde
* Iot gibi teknolojilerde verilerin anlamlı hale getirilmesinde kullanılır

# Veri Seti Hikayesi
Beyne bağlanan bir cihaz aracılığı ile toplanan sinyalleri ifade eden veri seti.

* subject : Verilerin toplandığı kişiler
* timepoint : Zaman noktaları
* event : Verinin toplanmasıyla ilgili olaylar
* region : Sinyalin toplandığı bölge
* signal : Gelen sinyal

```python
import seaborn as sns
fmri =sns.load_dataset("fmri")
df=fmri.copy()
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
df["timepoint"].describe()
```

```python
df["signal"].describe()
```

Her zaman noktasında eşit sinyal mi var?

```python
df.groupby("timepoint")["signal"].count()
```

Sinyaller tekil mi?

```python
df.groupby("signal").count()
```

```python
df.groupby("timepoint")["signal"].describe()
```

#Çizgi Grafik ve Çaprazlama

```python
sns.lineplot(x="timepoint",y="signal",data=df);
```

Somut çizgiler ortalama değerlere karşılık geliyor (56 değerin ortalaması)

Etrafındaki kısım std ile oluşturulmuş güven aralığı.

```python
sns.lineplot(x="timepoint",y="signal", hue="event",data=df);
```

Grafiğin pik yapan kısımlarda kategorik değişkenin etkisi görülmekte.

```python
df.columns
```

```python
sns.lineplot(x="timepoint",y="signal", hue="event",style="event",data=df);
```

Ortalama değerlerinin işaretlenmesi

```python
sns.lineplot(x="timepoint",y="signal", hue="event",style="event",markers=True,dashes=False,data=df);
```

Bölgelerinde grafiğe eklenmesi

```python
sns.lineplot(x="timepoint",y="signal",hue="region",style="event",data=df);
```

##Apple Hisseleri

```python
apple = yf.download("AAPL",start="2017-01-01", progress=False)
apple
```

#Veri Görselleştirme

```python
fig=go.Figure(data=[go.Candlestick(x=apple.index,
                                   open=apple['Open'],
                                   high=apple['High'],
                                   low=apple['Low'],
                                   close=apple['Close'])])
fig.show()
```
