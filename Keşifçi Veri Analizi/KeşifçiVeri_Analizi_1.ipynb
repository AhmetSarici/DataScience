```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns  
```

```python
data = pd.read_csv('/content/drive/My Drive/Colab Notebooks/Kaggle/pokemon.csv')
data.head()
```

# Line Plot

```python
plt.figure(figsize=(12,6))

plt.plot(data.index,data.Defense,color="red",label = 'Defence',linestyle = ':',linewidth=3) 
plt.plot(data.index,data.Attack,color="blue",label = 'Attack',linestyle = '-',linewidth=2)
plt.legend(loc='upper right')

plt.xlabel('x axis - Pokemonlar')              
plt.ylabel('y axis- DeÄŸer')
plt.title('Line Plot') 
plt.show()
```

```python
data.Speed.plot(kind = 'line', color = 'g',label = 'Speed',linewidth=1,alpha = 0.5,grid = True,linestyle = ':')
data.Defense.plot(color = 'r',label = 'Defense',linewidth=1, alpha = 0.5,grid = True,linestyle = '-.')

plt.legend(loc='upper right')     # legend = puts label into plot
plt.xlabel('x axis - Pokemonlar')              # label = name of label
plt.ylabel('y axis- DeÄŸer')
plt.title('Line Plot')            # title = title of plot
plt.show()
#x: pokemonlar
#y: hÄ±zlarÄ±
```

# Scatter Plot

```python
data.plot(kind='scatter', x='Attack', y='Defense',alpha = 0.5,color = 'red');
plt.xlabel('Attack');            
plt.ylabel('Defence');
plt.title('Attack Defense Scatter Plot'); 
```

```python
f=plt.figure(figsize=(12,6))

plt.scatter(data.Attack,data.Defense,s=100,color="orange",alpha = 0.5,
            edgecolors='black')


plt.xlabel("Atak")
plt.ylabel("Savunma")
plt.title("Pokemon Savunma ve Atak Durumu")
plt.show()
```
