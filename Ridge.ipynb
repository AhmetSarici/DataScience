## Model

```python
hit = pd.read_csv("Hitters.csv")
df = hit.copy()
df = df.dropna()
ms = pd.get_dummies(df[['League', 'Division', 'NewLeague']])
y = df["Salary"]

X_ = df.drop(['Salary', 'League', 'Division', 'NewLeague'], axis=1).astype('float64')
X = pd.concat([X_, dms[['League_N', 'Division_W', 'NewLeague_N']]], axis=1)

X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size=0.25, 
                                                    random_state=42)
```

```python
from sklearn.linear_model import Ridge
```

```python
ridge_model = Ridge(alpha = 0.1).fit(X_train, y_train)
```

```python
#alpha = λ
ridge_model
```

```python
# λ=1 iken katsayılar
ridge_model.coef_
```

```python
# Ceza için λ değerleri oluşturma işlemi
10**np.linspace(10,-2,100)*0.5 
```

```python
lambdalar = 10**np.linspace(10,-2,100)*0.5 

ridge_model = Ridge() # Model nesnesi
katsayilar = []

for i in lambdalar:
    ridge_model.set_params(alpha = i) # λ değeri gönderilecek
    ridge_model.fit(X_train, y_train) 
    katsayilar.append(ridge_model.coef_) #19 değişkenden dolayı 19 değer oluşacak 
    

    
ax = plt.gca()
ax.plot(lambdalar, katsayilar) 
ax.set_xscale('log') #x ekseni ayarları

plt.xlabel('Lambda(Alpha) Değerleri')
plt.ylabel('Katsayılar/Ağırlıklar')
plt.title('Düzenlileştirmenin Bir Fonksiyonu Olarak Ridge Katsayıları');
```

## Tahmin

```python
#  Tahmin edilen değer
y_pred = ridge_model.predict(X_test)
```

```python
# Test hatası
np.sqrt(mean_squared_error(y_test, y_pred))
```

## Model Tuning

```python
lambdalar = 10**np.linspace(10,-2,100)*0.5 
```

```python
lambdalar[0:5]
```

```python
#Hangi λ değeri ideal?
from sklearn.linear_model import RidgeCV
#Model Oluşturma
ridge_cv = RidgeCV(alphas = lambdalar, 
                   scoring = "neg_mean_squared_error",
                   normalize = True)#Normalleştirme işlemi
```

```python
#Model kurulumu
ridge_cv.fit(X_train, y_train)
```

```python
# CV optiumum landa değerini veren ridge metodu
ridge_cv.alpha_
```

```python
#Final (tune edilmiş) modeli
ridge_tuned = Ridge(alpha = ridge_cv.alpha_, 
                   normalize = True).fit(X_train,y_train)
```

```python
# Test hatası
np.sqrt(mean_squared_error(y_test, ridge_tuned.predict(X_test)))
```
